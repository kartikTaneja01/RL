```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
x = np.random.randint(0,50,500)  
```

```python
x
```

```python
y=x + np.random.normal(0,10,500)
y
```

```python
np.corrcoef(x,y)
```

```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.style.use('ggplot')
plt.scatter(x,y)
plt.show()
```

```python
import pandas as pd

df= pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b'] = df['a'] + np.random.normal(0,10,1000)
df['c'] = 100 - df['a'] + np.random.normal(0,5,1000)
df['d'] = np.random.normal(0,5,1000)
df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize=(5,5))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
